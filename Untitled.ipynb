```python
import random

r = random.randint(0,5)
print(r)
```

```python
for i in range(1000):
    r = random.randint(0,5)
    print(r)
```

```python
my_counter = dict()

for i in range(1000):
    r = random.randint(0,4)
    
    if r in my_counter.keys():
        my_counter[r] = my_counter[r] + 1
    else:
        my_counter[r] = 1
        
```

```python
import matplot.pyplot as plt
plt.bar(my_counter.keys(), my_counter.values())
```
